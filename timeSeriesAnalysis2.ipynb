```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#Load the dataset from YAHOO finance
ticker = 'tsla'
#data = yf.download(ticker, start='2022-01-01', end='2025-01-01')
data = yf.download(ticker, period='1mo', interval='5m') #1mo: 1 month, 5m:5 minutes
data.head()
```

```python
data.describe()
```

```python
data.tail() #MULTI-VARIENT DATA
```

```python
#Use the 'CLOSE' column for analysis
df= data[['Close']]
#Plotting
plt.figure(figsize=(15,3))
plt.plot(df['Close'], label='Close price')
plt.title('TESLA Price History')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.grid()
plt.legend()
plt.show()
```
