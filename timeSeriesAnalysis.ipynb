```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
series=pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/refs/heads/master/airline-passengers.csv')
series.head()
```

```python
series.info()
```

```python
series.describe()
```

```python
plt.figure(figsize=(12,4))
plt.plot(series['Passengers'])
plt.show()
```

```python
#Decompose the time series
result = sm.tsa.seasonal_decompose(series['Passengers'], model='additive', period=12)
```

```python
plt.figure(figsize=(12,10))
result.plot()
plt.show()
```
