# **1. Linear Activation Function**

```python
import numpy as np
import matplotlib.pyplot as plt

def linear(x):
    return x

def d_linear(x):
    return np.ones_like(x)

x = np.linspace(-5, 5, 400)
y = linear(x)
dy = d_linear(x)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(x, y, 'b', label="Linear Function")
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
plt.title("Linear Activation Function")
plt.legend()
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(x, dy, 'r', label="Derivative of Linear")
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
plt.title("Derivative of Linear Function")
plt.legend()
plt.grid()

plt.show()
```

# **2. Sigmoid Function**

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def d_sigmoid(x):
    s = sigmoid(x)
    return s * (1 - s)

x = np.linspace(-5, 5, 400)
y = sigmoid(x)
dy = d_sigmoid(x)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(x, y, 'b', label="Sigmoid Function")
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
plt.title("Sigmoid Activation Function")
plt.legend()
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(x, dy, 'r', label="Derivative of Sigmoid")
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
plt.title("Derivative of Sigmoid Function")
plt.legend()
plt.grid()

plt.show()
```

# **3. Hyperbolic Tangent (Tanh) Function**

```python
import numpy as np
import matplotlib.pyplot as plt

def tanh(x):
    return np.tanh(x)

def d_tanh(x):
    return 1 - np.tanh(x) ** 2

x = np.linspace(-5, 5, 400)
y = tanh(x)
dy = d_tanh(x)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(x, y, 'b', label="Tanh Function")
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
plt.title("Tanh Activation Function")
plt.legend()
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(x, dy, 'r', label="Derivative of Tanh")
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
plt.title("Derivative of Tanh Function")
plt.legend()
plt.grid()

plt.show()
```

# **4. ReLU (Rectified Linear Unit)**

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)

def d_relu(x):
    return np.where(x > 0, 1, 0)

x = np.linspace(-5, 5, 400)
y = relu(x)
dy = d_relu(x)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(x, y, 'b', label="ReLU Function")
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
plt.title("ReLU Activation Function")
plt.legend()
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(x, dy, 'r', label="Derivative of ReLU")
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
plt.title("Derivative of ReLU Function")
plt.legend()
plt.grid()

plt.show()
```

# **5. Leaky ReLU**

```python
import numpy as np
import matplotlib.pyplot as plt

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def d_leaky_relu(x, alpha=0.01):
    return np.where(x > 0, 1, alpha)

x = np.linspace(-5, 5, 400)
y = leaky_relu(x)
dy = d_leaky_relu(x)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(x, y, 'b', label="Leaky ReLU Function")
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
plt.title("Leaky ReLU Activation Function")
plt.legend()
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(x, dy, 'r', label="Derivative of Leaky ReLU")
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
plt.title("Derivative of Leaky ReLU Function")
plt.legend()
plt.grid()

plt.show()
```

# **6. ELU (Exponential Linear Unit)**

```python
import numpy as np
import matplotlib.pyplot as plt

def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

def d_elu(x, alpha=1.0):
    return np.where(x > 0, 1, elu(x, alpha) + alpha)

x = np.linspace(-5, 5, 400)
y = elu(x)
dy = d_elu(x)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(x, y, 'b', label="ELU Function")
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
plt.title("ELU Activation Function")
plt.legend()
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(x, dy, 'r', label="Derivative of ELU")
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
plt.title("Derivative of ELU Function")
plt.legend()
plt.grid()

plt.show()
```

# **7. GELU (Gaussian Error Linear Unit)**

```python
import numpy as np
import matplotlib.pyplot as plt

def gelu(x):
    return 0.5 * x * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * x**3)))

def d_gelu(x):
    sqrt_pi = np.sqrt(2 / np.pi)
    tanh_term = np.tanh(sqrt_pi * (x + 0.044715 * x**3))
    return 0.5 * (1 + tanh_term) + 0.5 * x * (1 - tanh_term**2) * sqrt_pi * (1 + 3 * 0.044715 * x**2)

x = np.linspace(-5, 5, 400)
y = gelu(x)
dy = d_gelu(x)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(x, y, 'b', label="GELU Function")
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
plt.title("GELU Activation Function")
plt.legend()
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(x, dy, 'r', label="Derivative of GELU")
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
plt.title("Derivative of GELU Function")
plt.legend()
plt.grid()

plt.show()
```

# **8. Parametric ReLU (PReLU)**

```python
import numpy as np
import matplotlib.pyplot as plt

def prelu(x, alpha=0.1):
    return np.where(x > 0, x, alpha * x)

def d_prelu(x, alpha=0.1):
    return np.where(x > 0, 1, alpha)

x = np.linspace(-5, 5, 400)
y = prelu(x)
dy = d_prelu(x)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(x, y, 'b', label="PReLU Function")
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
plt.title("Parametric ReLU Activation Function")
plt.legend()
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(x, dy, 'r', label="Derivative of PReLU")
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
plt.title("Derivative of Parametric ReLU Function")
plt.legend()
plt.grid()

plt.show()
```

# **9. Softmax Function**

```python
import numpy as np
import matplotlib.pyplot as plt

def softmax(x):
    exp_x = np.exp(x - np.max(x))  # Stability trick
    return exp_x / np.sum(exp_x)

def d_softmax(x):
    s = softmax(x)
    return s * (1 - s)

x = np.linspace(-5, 5, 400)
y = softmax(x)
dy = d_softmax(x)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(x, y, 'b', label="Softmax Function")
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
plt.title("Softmax Activation Function")
plt.legend()
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(x, dy, 'r', label="Derivative of Softmax")
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
plt.title("Derivative of Softmax Function")
plt.legend()
plt.grid()

plt.show()
```
